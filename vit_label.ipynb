```python
import json # import json module
```

```python
total={"straight":[],"curved":[],"static":[],"dynamic":[], "simple":[], "complex":[], "functional":[],"emotional":[],"natural":[],"artificial":[], "luxurious":[], "universal":[], "modern":[], "classic":[], "well-balanced":[], "tense":[], "achromatic":[],"chromatic":[], "cold":[], "warm":[], "hard":[], "soft":[]  }
```

```python
question_value={}
```

```python
with open('./newlabel/custom_question_train.json') as json_file:
    question = json.load(json_file)
with open('./newlabel/custom_annotation_train.json') as json_file:
    annotation = json.load(json_file) 
```

```python
for q in question['questions']:
    question_value[q['question_id']]=q['question'].split()[6]
```

```python
annotation["annotations"]
```

```python
annot
```

```python
for annot in annotation["annotations"]:
    path='/data/gist/ViT/dataset/image/resized/'+annot['image_id']
    save_answer=annot['answers'][0]['answer']
    total[question_value[annot['question_id']]].append([path, save_answer])
```

```python
total['warm']
```

```python
import csv
```

```python
#    CSV  
for key, items in total.items():
    filename = f"./dataset/label/train/{key}.csv"  #   (: straight.csv)
    with open(filename, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        for item in items:
            writer.writerow(item)  #   
```
